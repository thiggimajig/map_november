```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
import json
import requests
import geopandas
```

```python
url = (
    "https://www.submarinecablemap.com/api/v3/cable/cable-geo.json"
)

points = ("https://www.submarinecablemap.com/api/v3/landing-point/landing-point-geo.json")

all = ("https://www.submarinecablemap.com/api/v3/cable/all.json")
```

```python
m = folium.Map(
    location=[0, -30],
    tiles="cartodbpositron",
    zoom_start=2,
)

folium.GeoJson(url, name="cables").add_to(m)

folium.GeoJson(points, name="landing points").add_to(m)

# folium.TopoJson(    json.loads(requests.get(all).text),
#     "objects.all", name="all").add_to(m)


folium.LayerControl().add_to(m)
```
